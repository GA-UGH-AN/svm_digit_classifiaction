```python
import pandas as pd
import matplotlib.pyplot as plt

#Importing our dataset
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(2):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)
len(X_train)
```

```python
len(X_test)
```

```python
#Trying the rbf kernel
from sklearn.svm import SVC

model_rbf = SVC(kernel='rbf')
model_rbf.fit(X_train, y_train)
```

```python
#Trying the linear kernel

model_linear = SVC(kernel='linear')
model_linear.fit(X_train, y_train)
```

```python
#Trying the sigmoid kernel

model_sigmoid = SVC(kernel='sigmoid')
model_sigmoid.fit(X_train, y_train)
```

```python
#Checking the accuracy of our model with rbf kernel

model_rbf.score(X_test, y_test)
```

```python
#Checking the accuracy of our model with linear kernel

model_linear.score(X_test, y_test)
```

```python
#Checking the accuracy of our model with sigmoid kernel

model_sigmoid.score(X_test, y_test)
```

```python
digits.target[25] #Ground truth for these test values
```

```python
model_rbf.predict([digits.data[25]]) #Test prediction with rbf kernel
```

```python
model_linear.predict([digits.data[25]]) #Test prediction with linear kernel
```

```python
model_sigmoid.predict([digits.data[25]]) #Test prediction with sigmoid kernel
```

```python
#Using rbf kernel

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_prediction = model_rbf.predict(X_test)

accuracy = accuracy_score(y_test, y_prediction)
precision = precision_score(y_test, y_prediction, pos_label=1, average='weighted')
recall = recall_score(y_test, y_prediction, pos_label=1, average='weighted') 
f_score = f1_score(y_test, y_prediction, pos_label=1, average='weighted')

print("For Support Vector Machine (SVM) using RBF kernel")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F-score: {f_score}")
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_prediction)
print('RBF Kernel')
plt.figure(figsize = (6,3))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Prediction")
plt.ylabel("Truth")
```

```python
#Using Linear kernel

y_prediction = model_linear.predict(X_test)

accuracy = accuracy_score(y_test, y_prediction)
precision = precision_score(y_test, y_prediction, pos_label=1, average='weighted')
recall = recall_score(y_test, y_prediction, pos_label=1, average='weighted') 
f_score = f1_score(y_test, y_prediction, pos_label=1, average='weighted')

print("For Support Vector Machine (SVM) using Linear kernel")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F-score: {f_score}")
```

```python
cm = confusion_matrix(y_test, y_prediction)
print('Linear Kernel')
plt.figure(figsize = (6,3))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Prediction")
plt.ylabel("Truth")
```

```python
#Using Sigmoid kernel

y_prediction = model_sigmoid.predict(X_test)

accuracy = accuracy_score(y_test, y_prediction)
precision = precision_score(y_test, y_prediction, pos_label=1, average='weighted')
recall = recall_score(y_test, y_prediction, pos_label=1, average='weighted') 
f_score = f1_score(y_test, y_prediction, pos_label=1, average='weighted')

print("For Support Vector Machine (SVM) using Sigmoid kernel")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F-score: {f_score}")
```

```python
cm = confusion_matrix(y_test, y_prediction)
print('Sigmoid Kernel')
plt.figure(figsize = (6,3))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Prediction")
plt.ylabel("Truth")
```
